————预测————

```python
data1SortDummies=data1SortDummies[['入池水分（X3）', '入池酸度（X4）' ,'入池淀粉（X5）', '出池酒精度（Y5）']]
data1SortDummies
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(data1SortDummies[['入池水分（X3）', '入池酸度（X4）' ,'入池淀粉（X5）']],data1SortDummies[['出池酒精度（Y5）']],test_size=0.3)
```

```python
# 线性回归建模
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(train_x, train_y)
pred=regressor.predict(test_x)
mae = sum(abs(pred - test_y.values)/test_y.values) / len(pred)
print(mae)
```

```python
from sklearn.linear_model import LinearRegression, Ridge
import sklearn.pipeline as pl
import sklearn.preprocessing as sp
# regressor = Ridge(30, fit_intercept=True)


regressor = pl.make_pipeline(sp.PolynomialFeatures(degree=2), LinearRegression())
regressor.fit(train_x, train_y)
pred=regressor.predict(test_x)
mae = sum(abs(pred - test_y.values)/(test_y.values)) / len(pred)
print(mae)
```

```python
dataSSDJ.dropna(inplace=True, how='any')
X=dataSSDJ[['入池水分', '入池酸度' ,'入池淀粉']]
X.head()
y=dataSSDJ['出池酒精度']
y.head()
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)
# np.isnan(y_train).any()
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
model=linreg.fit(X_train, y_train)
# print(model)
# 训练后模型截距
print(linreg.intercept_)
# 训练后模型权重（相关系数）
# print(linreg.coef_)
print(list(zip(['入池水分', '入池酸度' ,'入池淀粉'], linreg.coef_)))
# y=4.914324464082131 - 0.026241220122483724∗入池水分 - 0.18677104689853138∗入池酸度 + 0.019211394573330404∗入池淀粉
y_pred=model.predict(X_test)
print(y_test.min(), y_test.max())
print(y_pred.min(), y_pred.max())

from sklearn import metrics
rmse=np.sqrt(metrics.mean_squared_error(y_test, y_pred))
r2=metrics.r2_score(y_test, y_pred)
rmse
# r2

# 相关系数
# dataSSDJ.corrwith(dataSSDJ['出池酒精度'])
# dataSSDJ.corr()
```

```python
# 多项式回归
from sklearn.preprocessing import PolynomialFeatures
quadratic = PolynomialFeatures(degree=2)
X_train_quad = quadratic.fit_transform(X_train)
X_test_quad = quadratic.fit_transform(X_test)
spr = LinearRegression()
spr.fit(X_train_quad, y_train)
# y_train_pred_quad = spr.predict(X_train_quad)
y_test_pred_quad = spr.predict(X_test_quad)

print(y_test.min(), y_test.max())
print(round(y_test_pred_quad.min(), 2), round(y_test_pred_quad.max(), 2))


rmse=np.sqrt(metrics.mean_squared_error(y_test, y_test_pred_quad))
r2=metrics.r2_score(y_test, y_test_pred_quad)
rmse
r2
```

```python
plt.figure()  
plt.scatter(range(len(y_test)),y_test_pred_quad,label="predict")  
plt.scatter(range(len(y_test)),y_test,label="test")  
plt.legend(loc="upper right") #显示图中的标签  
plt.xlabel("the number of sales")  
plt.ylabel('value of sales')  
plt.show() 
```
