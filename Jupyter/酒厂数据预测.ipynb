```python
f = open('E:\\jc\\出入池与出池酒精度_处理zh+天气.csv')
data=pd.read_csv(f, index_col='i')
data.index.name=None
data    # 2176
data.drop_duplicates(subset=None,keep='first',inplace=True)


mb=data.filter(regex=("天气.*"))
datab=data.drop(mb, axis=1)
tb=data.filter(regex=("最.*")).fillna(method='bfill')
tc=tb.fillna(method='ffill')
datab=pd.concat([data.loc[:, :'班次_白班'],tc], axis=1, join_axes=[tc.index])


# datam——只保留出池酒精度一个出池指标
colsm = [i for i in datab.columns if i not in ['出池水分', '出池酸度', '出池淀粉', '出池糖分','出池总酯', '己酸乙酯', '乙酸乙酯', '乳酸乙酯', '丁酸乙酯', '总酸', '乙醛', '甲醇', '正丙醇', '仲丁醇', '乙缩醛', '异丁醇', '正丁醇', '异戊醇', 'year', 'month', 'nian', 'chanliang', '高酯酒', '优级酒', '一级酒', '二级酒', '等级酒_一级', '等级酒_一降二', '等级酒_二级', '等级酒_优', '等级酒_优级', '等级酒_优降一']]
datam=datab[colsm]
datam.dropna(inplace=True, how='any')
datam[:].min()[:26]
datam[:].max()[:26]
datam.to_csv('E:\\jc\\Y出池酒精度不保留其他出池指标.csv', encoding="utf_8_sig")
datam


# datan——保留其他出池指标
colsn = [i for i in datab.columns if i not in ['year', 'month', 'nian', 'chanliang']]
datan=datab[colsn]
datan.dropna(inplace=True, how='any')
datan[:].min()[:26]
datan[:].max()[:26]
datan.to_csv('E:\\jc\\Y出池酒精度保留其他出池指标.csv', encoding="utf_8_sig")
datan


# data[:].min()[:51]
# data[:].max()[:51]

# datab[(datab['入池酸度']>1.2)&(datab['入池酸度']<2)]
```

```python
cols1 = [i for i in data.columns if i not in ['入池日期', '出池日期', '出池水分', '出池酸度', '出池淀粉', '出池糖分','出池总酯', '己酸乙酯', '乙酸乙酯', '乳酸乙酯', '丁酸乙酯', '总酸', '乙醛', '甲醇', '正丙醇', '仲丁醇', '乙缩醛', '异丁醇', '正丁醇', '异戊醇', 'year', 'month', 'nian', 'chanliang', '高酯酒', '优级酒', '一级酒', '二级酒', '等级酒_一级', '等级酒_一降二', '等级酒_二级', '等级酒_优', '等级酒_优级', '等级酒_优降一']]
data1=data[cols1]
data1
```

```python
from sklearn.cross_validation import train_test_split
data1=data1.loc[:,:'最低气温70天']
m=data1.filter(regex=("天气.*"))
data1=data1.drop(m, axis=1)
t=data1.filter(regex=("最.*")).fillna(method='bfill')
data1=pd.concat([data1.loc[:, :'班次_白班'],t], axis=1, join_axes=[t.index])
data1.dropna(inplace=True, how='any')

X=data1.drop('出池酒精度', axis=1)
y=data1['出池酒精度']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
data1
```

————回归预测————

```python
# 多项式回归
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
quadratic = PolynomialFeatures(degree=2)
X_train_quad = quadratic.fit_transform(X_train)
X_test_quad = quadratic.fit_transform(X_test)
spr = LinearRegression()
spr.fit(X_train_quad, y_train)
# y_train_pred_quad = spr.predict(X_train_quad)
y_test_pred_quad = spr.predict(X_test_quad)

mse=metrics.mean_squared_error(y_test, y_test_pred_quad)
print('mse: %.4f' % mse)
r2=metrics.r2_score(y_test, y_test_pred_quad)
# r2
mae=metrics.mean_absolute_error(y_test, y_test_pred_quad)
# mae

sum(abs(y_test_pred_quad - y_test)/(y_test)) / len(y_test_pred_quad)
```

```python
# svm
from sklearn.svm import SVR

# standardization特征缩放—数据规范化，使特征的范围具有可比性
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# scaler.fit(X_train)
# X_train = scaler.transform(X_train)
# X_test = scaler.transform(X_test)

svr = SVR(kernel = 'rbf')
svr.fit(X_train, y_train)
y_predict = svr.predict(X_test)
np.hstack((y_test.values.reshape(-1, 1), y_predict.reshape(-1, 1)))

mse=metrics.mean_squared_error(y_test, y_predict)
print('mse: %.4f' % mse)

sum(abs(y_predict - y_test)/(y_test)) / len(y_predict)
```

```python
# 随机森林
from sklearn.ensemble import RandomForestRegressor

# 采用DictVectorizer进行特征向量化—提升代码运行效率
# from sklearn.feature_extraction import DictVectorizer
# dict_vec = DictVectorizer(sparse=False)     # sparse=False意思是不产生稀疏矩阵
# X_train = dict_vec.fit_transform(X_train.to_dict(orient='record'))
# X_test = dict_vec.transform(X_test.to_dict(orient='record')

rfr = RandomForestRegressor()      # n_estimators=20, max_depth=30
rfr.fit(X_train, y_train)

# 平均不纯度减少 mean decrease impurity
# print(sorted(zip(map(lambda x: round(x, 4), rfr.feature_importances_), X.columns), reverse=True))

rfr_y_predict = rfr.predict(X_test)

mse=metrics.mean_squared_error(y_test, rfr_y_predict)
print('mse: %.4f' % mse)

sum(abs(rfr_y_predict - y_test)/(y_test)) / len(rfr_y_predict)
```

```python
# 决策树
from sklearn import tree
dtr=tree.DecisionTreeRegressor()
dtr.fit(X_train, y_train)
dtr_y_predict = dtr.predict(X_test)

mse=metrics.mean_squared_error(y_test, dtr_y_predict)
print('mse: %.4f' % mse)

sum(abs(dtr_y_predict - y_test)/(y_test)) / len(dtr_y_predict)
```
